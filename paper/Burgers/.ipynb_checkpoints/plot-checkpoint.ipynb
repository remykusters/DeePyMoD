```python
spline_0 = np.array([np.mean(row.deriv_error) for idx, row in df_spline.iterrows()])
spline_2 = np.array([np.mean(row.deriv_error) for idx, row in df_spline_noisy.iterrows()])
dx_spline = np.array([np.diff(row.x)[0] for idx, row in df_spline.iterrows()])

deepmod_0 = np.array([np.mean(row.deriv_error) for idx, row in df_deepmod_noiseless.iterrows()])
deepmod_2 = np.array([np.mean(row.deriv_error) for idx, row in df_deepmod_noisy.iterrows()])

dx_deepmod = np.array([np.diff(row.x)[0] for idx, row in df_deepmod_noiseless.iterrows()])
reg = LinearRegression()
```

```python
reg.fit(np.log(dx_spline[:, None]), np.log(spline_0))
plt.loglog(dx_spline, np.exp(reg.intercept_) * dx_spline **(reg.coef_), 'k-', label='Spline, 0% noise', linewidth=1.5)
reg.fit(np.log(dx_spline[:, None]), np.log(spline_2))
plt.loglog(dx_spline, np.exp(reg.intercept_) * dx_spline **(reg.coef_), 'g--', label='Spline, 20% noise', linewidth=1.5)
reg.fit(np.log(dx_deepmod[:, None]), np.log(deepmod_0))
plt.loglog(dx_deepmod, np.exp(reg.intercept_) * dx_deepmod **(reg.coef_), 'r-', label='Deepmod, 0% noise', linewidth=1.5)
reg.fit(np.log(dx_deepmod[:, None]), np.log(deepmod_2))
plt.loglog(dx_deepmod, np.exp(reg.intercept_) * dx_deepmod **(reg.coef_), 'b--', label='Deepmod, 20% noise', linewidth=1.5)



for idx, row in df_spline.iterrows():
    if row.correct == True:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'ko', markersize=6)
    else:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'kv', markersize=6)



for idx, row in df_spline_noisy.iterrows():
    if row.correct == True:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'go', markersize=6)
    else:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'gv', markersize=6)


for idx, row in df_deepmod_noiseless.iterrows():
    if row.correct == True:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'ro', markersize=6)
    else:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'rv', markersize=6)



for idx, row in df_deepmod_noisy.iterrows():
    if row.correct == True:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'bo', markersize=6)
    else:
        plt.loglog(np.diff(row.x)[0], np.mean(row.deriv_error), 'bv', markersize=6)


plt.legend()       
plt.axvline(np.sqrt(4*0.25*0.1), color='k', linestyle='--')
plt.xlabel(r'$\Delta x$')
plt.ylabel('Relative error')
plt.savefig('error_spacing.pdf')
```
